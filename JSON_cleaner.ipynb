```python
with open('hackupc2023_restbai__dataset.json', 'r') as f:
    raw_dataset = json.loads(f.read())

reduced_dataset = itertools.islice(raw_dataset.copy().items(), 500000)
raw_dataset = reduced_dataset
```

```python
df = pd.DataFrame(columns=['summary', 'city', 'neighborhood', 'region', 'price', 'square_meters', 'bedrooms', 'bathrooms','property_rating', 'kitchen_rating', 'bathroom_rating', 'interior_rating'])

for key, value in raw_dataset:
    property_dict = {}
    property_dict['summary'] = value['summary']
    property_dict['city'] = value['city']
    property_dict['neighborhood'] = value['neighborhood']
    property_dict['region'] = value['region']
    property_dict['price'] = value['price']
    property_dict['square_meters'] = value['square_meters']
    property_dict['bedrooms'] = value['bedrooms']
    property_dict['bathrooms'] = value['bathrooms']
    if 'image_data' in value and 'r1r6' in value['image_data']:
        ratings = value['image_data']['r1r6']
        property_dict['property_rating'] = ratings.get('property', None)
        property_dict['kitchen_rating'] = ratings.get('kitchen', None)
        property_dict['bathroom_rating'] = ratings.get('bathroom', None)
        property_dict['interior_rating'] = ratings.get('interior', None)
    else:
        property_dict['property_rating'] = None
        property_dict['kitchen_rating'] = None
        property_dict['bathroom_rating'] = None
        property_dict['interior_rating'] = None
    df = df.append(property_dict, ignore_index=True)
```

```python
df = df.drop('summary', axis=1)
df = df.dropna()
df.head()
```
